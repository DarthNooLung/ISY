```python
import numpy as np
import cv2
import socketio
import base64
```

```python
imdata = 0

sio = socketio.Client()
```

```python
"""
@는 데코레이터임
https://bluese05.tistory.com/30 설명
"""

@sio.event
def streaming(data):
    data = cv2.imdecode(np.frombuffer(base64.b64decode(data), dtype='uint8'), cv2.IMREAD_COLOR)
    print(data)
    global imdata
    imdata = data
```

```python
sio.connect('http://localhost:746')

while True:
    sio.on('streaming', streaming)
    cv2.imshow('Person Detect', imdata)
    if cv2.waitKey(1) == 13:
        break
cv2.destroyAllWindows()
sio.disconnect()
```
